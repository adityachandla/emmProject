```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def analyze(df):
    print("Correlation=", np.corrcoef(df['Price'], df['Landsize'])[0][1])
    print("Data frame size is ", len(df))
    plt.ylabel("Price")
    plt.xlabel("Landsize")
    plt.scatter(df['Price'], df['Landsize'], alpha=0.6)
```

```python
housing_df = pd.read_csv("./melb_data.csv")
analyze(housing_df)
```

### Car >= 1 Suburb='Pascoe Vale' Room <= 2

```python
condition_rows = housing_df.loc[housing_df['Car'] >= 1].loc[housing_df['Suburb'] == 'Pascoe Vale'].loc[housing_df['Rooms'] >= 2]
analyze(condition_rows)
```

# Subgroup within subgroup
## houseType='h' and Cars <= 5 is exceptional within Suburb='Richmond'

```python
richmond_rows = housing_df.loc[housing_df['Suburb'] == 'Richmond']
analyze(richmond_rows)
```

```python
richmond_subgroup = richmond_rows.loc[richmond_rows['Type'] == 'h'].loc[richmond_rows['Car'] <= 5]
analyze(richmond_subgroup)
```

## Suburb=Brigton & Cars <= 5 is Exceptional within Rooms >= 3

```python
three_rooms = housing_df.loc[housing_df['Rooms'] >= 3]
analyze(three_rooms)
```

```python
brigton = three_rooms.loc[three_rooms["Suburb"] == "Brighton"].loc[three_rooms["Car"] <= 5]
analyze(brigton)
just_brigton = housing_df.loc[housing_df["Suburb"] == "Brighton"].loc[housing_df["Car"] <= 5]
analyze(just_brigton)
```

## Housetype=h and Cars >= 1 is exceptional in Rooms <= 2

```python
less_than_two = housing_df.loc[housing_df["Rooms"] <= 2]
analyze(less_than_two)
```

```python
houseType_h = less_than_two.loc[less_than_two['Type'] == 'h'].loc[less_than_two['Car'] >= 1]
analyze(houseType_h)
```

## Suburb=Hawthorn and Cars <= 5 is exceptional within HouseType=h

```python
h_houseType = housing_df.loc[housing_df['Type']=='h']
analyze(h_houseType)
```

```python
hawthorn = h_houseType.loc[h_houseType['Car'] <= 5].loc[h_houseType['Suburb'] == 'Hawthorn']
analyze(hawthorn)
```

## Car <= 3 and SellerName = Jellis exceptional within HouseType = h

```python
analyze(h_houseType)
```

```python
jellis = h_houseType.loc[h_houseType['SellerG'] == 'Jellis'].loc[h_houseType['Car'] <= 3]
analyze(jellis)
just_jellis = housing_df.loc[housing_df['SellerG'] == 'Jellis'].loc[housing_df['Car'] <= 3]
analyze(just_jellis)
```

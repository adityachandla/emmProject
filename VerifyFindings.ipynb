```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
housing_df = pd.read_csv("./melb_data.csv")
price_all = housing_df['Price'].to_numpy()
landsize_all = housing_df['Landsize'].to_numpy()
print("Actual correlation between landsize and price ", np.corrcoef(price_all, landsize_all)[0][1])
```

```python
plt.scatter(price_all, landsize_all)
plt.xlabel("Landsize")
plt.ylabel("Price")
```

### Car <= 3 Suburb='Reservoir' Room <= 2

```python
condition_rows = housing_df.loc[housing_df['Car'] <= 3].loc[housing_df['Suburb'] == 'Reservoir'].loc[housing_df['Rooms'] <= 2]
condition_price = condition_rows['Price']
condition_landsize = condition_rows['Landsize']
plt.scatter(condition_price, condition_landsize)
print("Correlation for this group is ", np.corrcoef(condition_price, condition_landsize)[0][1])
```

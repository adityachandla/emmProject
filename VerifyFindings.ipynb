```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
housing_df = pd.read_csv("./melb_data.csv")
price_all = housing_df['Price'].to_numpy()
landsize_all = housing_df['Landsize'].to_numpy()
print("Actual correlation between landsize and price ", np.corrcoef(price_all, landsize_all)[0][1])

print(housing_df.loc[housing_df['Car'] == np.nan])
```

```python
plt.scatter(price_all, landsize_all)
```

### Car >= 1 Suburb='Pascoe Vale' Room <= 2

```python
condition_rows = housing_df.loc[housing_df['Car'] >= 1].loc[housing_df['Suburb'] == 'Pascoe Vale'].loc[housing_df['Rooms'] >= 2]
condition_price = condition_rows['Price']
condition_landsize = condition_rows['Landsize']
plt.scatter(condition_price, condition_landsize)
print("Correlation for this group is ", np.corrcoef(condition_price, condition_landsize)[0][1])
```

## Subgroup within subgroup
The subgroup with houseType='h' and Cars <= 5 is exceptional within Suburb='Richmond'

```python
richmond_rows = housing_df.loc[housing_df['Suburb'] == 'Richmond']
richmond_price = richmond_rows['Price']
richmond_landsize = richmond_rows['Landsize']
plt.scatter(richmond_price, richmond_landsize)
print("Correlation for houses in richmond is ", np.corrcoef(richmond_price, richmond_landsize)[0][1])
```

```python
richmond_subgroup = richmond_rows.loc[richmond_rows['Type'] == 'h'].loc[richmond_rows['Car'] <= 5]
subgroup_price = richmond_subgroup['Price']
subgroup_landsize = richmond_subgroup['Landsize']
plt.scatter(subgroup_price, subgroup_landsize)
print("Correlation for the subgroup is ", np.corrcoef(subgroup_price, subgroup_landsize)[0][1])
```
